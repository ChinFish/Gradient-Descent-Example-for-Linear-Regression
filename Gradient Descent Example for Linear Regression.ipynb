```python
#y=w0 + w1*x1 + w2*x2 + w3*x3 + w4*x4 + w5*x5
#E = 1/n*(y-(w0 + w1*x1 + w2*x2 + w3*x3 + w4*x4 + w5*x5))**2
#微分後:E=-2/n*xi(yi-(w0 + w1*x1 + w2*x2 + w3*x3 + w4*x4 + w5*x5))
def compute_error_for_line_given_points(w0, w1, w2, w3, w4, w5, points):
    '''計算誤差'''
    totalError = 0
    for i in range(0, len(points)):
        x1 = points[i][0]
        x2 = points[i][1]
        x3 = points[i][2]
        x4 = points[i][3]
        x5 = points[i][4]
        y = points[i][5]
        totalError+=abs(y-(w0 + w1*x1 + w2*x2 + w3*x3 + w4*x4 + w5*x5))
        #totalError += (y - (w0 + w1*x1 + w2*x2 + w3*x3 + w4*x4 + w5*x5))**2
        #print(totalError)
    #return totalError / float(len(points))
    return totalError

def step_gradient(w0_current,w1_current,w2_current,w3_current,w4_current,w5_current, points, learningRate):
    w0_gradient = 0
    w1_gradient = 0
    w2_gradient = 0
    w3_gradient = 0
    w4_gradient = 0
    w5_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x1 = points[i][0]
        x2 = points[i][1]
        x3 = points[i][2]
        x4 = points[i][3]
        x5 = points[i][4]
        y = points[i][5]
        y_pred = w0_current + (w1_current*x1)+(w2_current*x2)+(w3_current*x3)+(w4_current*x4)+(w5_current*x5)
        w0_gradient += -(2/N) * (y-y_pred)
        w1_gradient += -(2/N) * x1 * (y-y_pred)
        w2_gradient += -(2/N) * x2 * (y-y_pred)
        w3_gradient += -(2/N) * x3 * (y-y_pred)
        w4_gradient += -(2/N) * x4 * (y-y_pred)
        w5_gradient += -(2/N) * x5 * (y-y_pred)

    new_w0 = w0_current - (learningRate * w0_gradient)
    new_w1 = w1_current - (learningRate * w1_gradient)
    new_w2 = w2_current - (learningRate * w2_gradient)
    new_w3 = w3_current - (learningRate * w3_gradient)
    new_w4 = w4_current - (learningRate * w4_gradient)
    new_w5 = w5_current - (learningRate * w5_gradient)
    return [new_w0,new_w1,new_w2,new_w3,new_w4,new_w5]

def gradient_descent_runner(points, starting_w0, starting_w1, starting_w2, starting_w3, starting_w4, starting_w5, learning_rate, num_iterations):
    w0 = starting_w0
    w1 = starting_w1
    w2 = starting_w2
    w3 = starting_w3
    w4 = starting_w4
    w5 = starting_w5
    maxn=0
    minn=99999
    num = compute_error_for_line_given_points(w0, w1, w2, w3, w4, w5, points)
    for i in range(num_iterations):
        w0,w1,w2,w3,w4,w5 = step_gradient(w0, w1, w2, w3, w4, w5, points, learning_rate)
        #print('w0',w0,'w1',w1,'w2',w2,'w3',w3,'w4',w4,'w5',w5)
        num = compute_error_for_line_given_points(w0, w1, w2, w3, w4, w5, points)
        if(maxn<num):
            maxn = num
        if(minn>num):
            minn = num
    #print('max',maxn,'min',minn)
    return [w0,w1,w2,w3,w4,w5]

def run():
    points = ([[149, 118, 15.8, 107, 4.3, 221.16],[214, 89, 37.7, 79, 4, 254.79],[216, 49, 38.6, 50, 5.3, 174.77],
               [183, 105, 29, 109, 2.6, 238.56],[214, 37, 20.2, 64, 9.4, 144.55],[86, 84, 23.5, 46, 2, 185.14],
               [214, 61, 22.2, 108, 8.1, 174.48],[137, 97, 43.6, 54, 0.9, 242.04],[138, 32, 27.1, 101, 4.7, 124.23],
               [104, 94, 34.6, 48, 0.9, 200.39],[176, 117, 43.5, 54, 2.7, 281.39],[199, 120, 45, 72, 9.4, 295.52],
               [167, 68, 39.3, 50, 9.8, 206.13],[105, 64, 13.4, 91, 7.5, 143.14],[83, 28, 27.4, 82, 1.8, 119.84],
               [101, 57, 21.8, 101, 6.3, 147.56],[197, 40, 44.6, 63, 10, 185.70],[193, 83, 33.5, 97, 5.5, 208.00],
               [83, 35, 36.4, 104, 5.3, 140.65],[83, 111, 27.2, 96, 1, 226.17]])
    #print(type(points))
    learning_rate = 0.00002
    initial_w0 = 0
    initial_w1 = 0
    initial_w2 = 0
    initial_w3 = 0
    initial_w4 = 0
    initial_w5 = 0
    
    num_iterations = 10000
    print ("一開始的w0 = {0}, w1 = {1}, w2 = {2}, w3 = {3}, w4 = {4}, w5 = {5}, 總誤差 = {6}".format(initial_w0, initial_w1, initial_w2,initial_w3,initial_w4,initial_w5,
                                                                            compute_error_for_line_given_points(initial_w0, initial_w1, initial_w2,initial_w3,initial_w4,initial_w5, points)))
    [w0, w1, w2, w3, w4, w5] = gradient_descent_runner(points, initial_w0, initial_w1, initial_w2,initial_w3,initial_w4,initial_w5, learning_rate, num_iterations)
    
    print ("完成 {0} 次運算後 w0 = {1:.4f}, w1 = {2:.4f},w2 = {3:.4f},w3 = {4:.4f},w4 = {5:.4f},w5 = {6:.4f}, 總誤差 = {7}".format(num_iterations, w0, w1, w2, w3, w4, w5, 
                                                                                            compute_error_for_line_given_points(w0, w1, w2, w3, w4, w5, points)))
    print('方程式:y={0:.2f} + {1:.2f}x1 + {2:.2f}x2 + {3:.2f}x3 + {4:.2f}x4 + {5:.2f}x5\n'.format(w0,w1,w2,w3,w4,w5))
    
    maxn = 0
    minn = 9999
    for i in points:
        y_pre = w0 + w1*i[0] + w2*i[1] + w3*i[2] + w4*i[3] + w5*i[4]
        a = abs(y_pre - i[5])
        if(maxn<a):
            maxn = a
        if(minn>a):
            minn = a
        print('原始的y:',i[5],'後來的y',y_pre)
    print('max',maxn,'min',minn)
        
    print('總差距:',compute_error_for_line_given_points(w0, w1, w2, w3, w4, w5, points))
    print('平均誤差:',compute_error_for_line_given_points(w0, w1, w2, w3, w4, w5, points)/len(points))
    #print(len(points))
if __name__ == '__main__':
    run()
```
